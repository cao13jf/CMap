```python
# add the new cell if the cell did not at the name_dictionary_cmap.csv
import glob
import os
from data.generate_number_dictionary import add_number_dict


CD_folder = r"./dataset/run/CDFiles"
nuc_files = sorted(glob.glob(os.path.join(CD_folder, "*.csv")))

for idx, nuc_file in enumerate(nuc_files): # go through all cd files, one cd file for one embryo
    add_number_dict(nuc_file, max_time=1000) # the max time for your data
```

### Compose the 2D raw membrane channel tiff images into 3D niigz volumes

```python
# configuration
config = dict(num_slice=92, # the depth, the z axia, how many slices along z
                  embryo_names=['200109plc1p1','200113plc1p2'], # you can give a list if you have many embryos, remember to group them under ./dataset/run/{embryo name}
                  max_times = [180,187], # the max time, should match the embryo
                  xy_resolution = 0.09,
                  z_resolution = 0.42,
                  # the z dimension is calculated by: 92  *   0.42/0.09  *  356/712; (2d height / 3d height)
                  out_size = [256, 356, 214], # todo: need to be calculated with the vertical image amount
                  raw_folder=r'./dataset/run/',
                  target_folder=r"./dataset/run/",
                  save_nuc = False, # group the nucleus channel images to 3d niigz file
                  save_memb = True, # group the membrane channel images to 3d niigz file
                  lineage_file = True,# group the cell position from cd file in to niigz for seeded watershed instance segmentation
                  name_dictionary = r"./dataset/name_dictionary.csv"
                  )


num_slice = config["num_slice"]
embryo_names = config["embryo_names"]
max_times = config["max_times"]
xy_res = config["xy_resolution"]
z_res = config["z_resolution"]
out_size = config["out_size"]
raw_folder = config["raw_folder"]
stack_folder = config["target_folder"]
is_save_nuc_channel = config["save_nuc"]
is_save_memb_channel = config["save_memb"]
is_save_seg_cell_with_cd_file=config["lineage_file"]
name_dictionary_path=config['name_dictionary']
```

```python
from PIL import Image
import numpy as np
import multiprocessing as mp


# preparation before composing images

# get output size
raw_memb_files = glob.glob(os.path.join(raw_folder, embryo_names[0], "tifR", "*.tif"))
print('length of the raw membrane/nucleus slices', len(raw_memb_files))
raw_size = list(np.asarray(Image.open(raw_memb_files[0])).shape) + [int(num_slice * z_res / xy_res)]
out_res = [res * in_scale / out_scale for res, in_scale, out_scale in zip([xy_res, xy_res, xy_res], raw_size, out_size)]
print('composing output shape resolution',out_res)

# multiprocessing
mpPool = mp.Pool(mp.cpu_count() - 1)
```


### Compose the membrane and nucleus; generate the segNuc according to CD{embryo name}.csv file  

```python
from tqdm import tqdm
import pandas as pd

from data.preprocess_utils import save_nuc_seg,stack_memb_slices,stack_nuc_slices
from data.data_structure import read_cd_file

for idx_embryo,embryo_name in enumerate(embryo_names):

    # get lineage file successfully?
    if is_save_seg_cell_with_cd_file:
        lineage_file_path = os.path.join(raw_folder, 'CDFiles', "CD{}.csv".format(embryo_name))
    else:
        lineage_file_path = None

    # ------------------Not provided here--- save nucleus-------------------------
    if is_save_nuc_channel:
        origin_folder = os.path.join(raw_folder, embryo_name, "tif")
        target_folder = os.path.join(stack_folder, embryo_name, "RawNuc")
        if not os.path.isdir(target_folder):
            os.makedirs(target_folder)

        configs = []
        for tp in range(1, max_times[idx_embryo] + 1):
            configs.append((origin_folder, target_folder, embryo_name, tp, out_size, num_slice, out_res))

        for idx, _ in enumerate(tqdm(mpPool.imap_unordered(stack_nuc_slices, configs), total=len(configs),
                                     desc="Stacking nucleus of {}".format(embryo_name))):
            pass

    # ---------------------- save membrane --------------------------
    if is_save_memb_channel:
        origin_folder = os.path.join(raw_folder, embryo_name, "tifR")
        target_folder = os.path.join(stack_folder, embryo_name, "RawMemb")
        if not os.path.isdir(target_folder):
            os.makedirs(target_folder)

        configs = []
        for tp in range(1, max_times[idx_embryo] + 1):
            configs.append((origin_folder, target_folder, embryo_name, tp, out_size, num_slice, out_res))
            # stack_memb_slices(configs[0])
        for idx, _ in enumerate(tqdm(mpPool.imap_unordered(stack_memb_slices, configs), total=len(configs),
                                     desc="Stacking membrane of {}".format(embryo_name))):
            pass

    # save nucleus
    if is_save_seg_cell_with_cd_file:
        assert lineage_file_path is not None
        target_folder = os.path.join(stack_folder, embryo_name, "SegNuc")
        if not os.path.isdir(target_folder):
            os.makedirs(target_folder)
        pd_lineage = read_cd_file(lineage_file_path)

        label_name_dict = pd.read_csv(name_dictionary_path, index_col=0).to_dict()['0']
        name_label_dict = {value: key for key, value in label_name_dict.items()}

        configs = []
        for tp in range(1, max_times[idx_embryo] + 1):
            configs.append((embryo_name, name_label_dict, pd_lineage, tp, raw_size, out_size, out_res,
                            xy_res / z_res, target_folder))
            # save_nuc_seg(configs[0])
        for idx, _ in enumerate(tqdm(mpPool.imap_unordered(save_nuc_seg, configs), total=len(configs),
                                     desc="Constructing segNuc(nucleus location) of {}".format(embryo_name))):
            # TODO: Process Name: `3/3 Construct nucleus location`; Current status: `idx`; Final status: max_time
            pass
#         shutil.copy(lineage_file_path, os.path.join(stack_folder, embryo_name))
```
